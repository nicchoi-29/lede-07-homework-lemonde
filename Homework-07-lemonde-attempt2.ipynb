```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np 
```

```python
response = requests.get("https://www.lemonde.fr/en/")
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
stories = doc.find_all("div", class_="article")
bottom_stories = doc.find_all("div", class_="area.area--section")

#print(doc.prettify())

#articles = doc.find_all("p", class_="article__title")
#subhead = doc.find_all("p", class_="article__desc")

rows = []

#ARTICLES
for story in stories:
    print("----")
    row = {}

    #TITLES
    try: 
        row["title"] = story.select_one("p.article__title," 
                                        "div.article__title," 
                                        "h1.article__title," 
                                        "a.data-zone," 
                                        "p.article__title--inline," 
                                        "div.article__title.mb0," 
                                        "li.old__article-related-default-item").text.strip()
   
    except:
        print("Headline not found")

        
    #LINK
    try:
        row['href'] = story.select_one('a')['href']
    except:
        print("Couldn't find a link")


    #SUBHEAD
    try:
        row['subhead'] = story.select_one("p.article__desc").text.strip()
    except:
        print("Couldn't find a subhead")



    #BYLINE
    try:
        row['byline'] = story.select_one("p.article__byline,"
                                        "div.article__author-name").text.strip()
    except:
        print("Couldn't find a byline")

   

    #ARTICLE TYPE
    try:
        row['type'] = story.select_one("p.article__type," "div.article__type").text.strip()
    except:
        print("Couldn't find a type")


    #PREMIUM?
    try:
        row['premium'] = story.select_one(".icon__premium").text.strip()
    
    except:
        pass


    # IMAGE URL
    try:
        row['image'] = story.select_one('img')['src']
    
    except:
         print("Couldn't find an image")
        
    print(row)
    rows.append(row)

####-----ARTICLES FOR LOWER PAGE----####

#ARTICLE TITLES 
for bottom_story in bottom_stories:
    print("----")
    row = {}
    
#     #TITLES
    try:
        row["title"] = story.select_one("p.article__title," 
                                        "div.article__title," 
                                        "h1.article__title," 
                                        "a.data-zone,"
                                        "p.article__title--inline," 
                                        "div.article__title.mb0," 
                                        "li.old__article-related-default-item").text.strip()
   
    except:
        print("Headline not found")
        

#     #LINK
    try:
        # Find me a media__link OR a reel_link
        row['href'] = bottom_story.select_one('a')['href']
    except:
        print("Couldn't find a link")


    #NO SUBHEADS

    #BYLINE
    try:
        row['byline'] = bottom_story.select_one("p.article__byline,"
                                        "div.article__author-name").text.strip()
    except:
        print("Couldn't find a byline")
    
    #ARTICLE TYPE
    try:
        row['type'] = bottom_story.select_one("p.article__type," "div.article__type").text.strip()
    except:
        print("Couldn't find a type")


    # #PREMIUM?
    try:
        row['premium'] = bottom_story.select_one(".icon__premium").text.strip()
    except:
        pass



    # #IMAGE URL
    try:
        row['image'] = bottom_story.select_one('img')['src']
    
    except:
         print("Couldn't find an image")
        
        
    print(row)
    if row.get("title") and row.get("href"):
        rows.append(row)
```

```python
df = pd.DataFrame(rows)
```
